```python
import numpy as np
import pandas as pd
```

```python
df_ratings = pd.read_csv('/content/ratings.csv',header = None, sep = '::',names = ["UserId","MovieId","Rating","Timestamp"])
df_ratings
```

```python
df_user = pd.read_csv('/content/users.csv',header = None, sep = '::', names = ["UserId","Gender","Age","Occupation","ZipCode"])
df_user
```

```python
df_movies = pd.read_csv('/content/movies.csv',encoding='latin-1',on_bad_lines='skip',header=None, names= ["MovieId","Title","Genres"],sep= '::')
df_movies
```

Creating Pivot Table for gaining more insights and to apply Collaborative Learning

```python
pivot_table = df_ratings.pivot_table(values='Rating',columns='UserId',index='MovieId').fillna(0)

pivot_table.head()
```

```python
print(pivot_table.shape)
```

```python
from sklearn.model_selection import train_test_split

train_data, test_data = train_test_split(pivot_table, test_size=0.2, random_state=2023)
```

```python
train_data
```

```python
test_data
```

For collaborative filtering, models like matrix factorization, nearest neighbors, or deep learning methods are more common and effective.

I will be using Nearest Neighbors model in this code.

```python
from sklearn.neighbors import NearestNeighbors

model = NearestNeighbors(metric='cosine')
model.fit(train_data)
```

```python
# Function to predict ratings for the test set
def predict_ratings(user_index):
    distances, indices = model.kneighbors(test_data.iloc[user_index, :].values.reshape(1, -1), n_neighbors=5)
    neighbor_ratings = train_data.iloc[indices[0], :]
    predicted_ratings = np.mean(neighbor_ratings, axis=0)
    return predicted_ratings
```

```python
predicted_ratings = [predict_ratings(user_index) for user_index in range(len(test_data))]
```

```python
actual_ratings_flat = test_data.values.flatten()
predicted_ratings_flat = np.concatenate(predicted_ratings)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import confusion_matrix
import math
```

```python
rmse = math.sqrt(mean_squared_error(actual_ratings_flat, predicted_ratings_flat))
print(f"Root Mean Squared Error (RMSE): {rmse}")
```

This value indicates that, on average, our model's predictions are off by approximately 0.7107 units on a scale of 1 to 5.

Lower RMSE values indicates better performance of the model.
So getting the RMSE value equal to 0.7107 indicates better performance of Nearest Neighbours model.

```python
# Converting ratings to binary for confusion matrix
# (If rating >= 3, considering it as positive)
actual_binary = (actual_ratings_flat >= 3).astype(int)
predicted_binary = (predicted_ratings_flat >= 3).astype(int)
```

```python
conf_matrix = confusion_matrix(actual_binary, predicted_binary)
print("Confusion Matrix:")
print(conf_matrix)
```

```python
class Recommender:
    def __init__(self):
        # To stroe the list of recommendation movies history.
        self.hist = []
        self.ishist = False # For checking whether history is empty or not.

    # This method will recommend movies based on a movie that is passed as the parameter.
    # This is for those times when you know you have watched a particular movie and want to watch similar movies.
    # I am setting the function to recommend 5 movies
    def recommend_on_movie(self,movie,n_reccomend = 5):
        self.ishist = True
        movie_id = int(df_movies[df_movies['Title']==movie]['MovieId'])
        self.hist.append(movie_id) #Adding the movie id in list

        distance,neighbors = model.kneighbors([pivot_table.loc[movie_id]],n_neighbors=n_reccomend+1)
        movie_ids = [pivot_table.iloc[i].name for i in neighbors[0]]
        recommendations = [str(df_movies[df_movies['MovieId']==mid]['Title']).split('\n')[0].split('  ')[-1] for mid in movie_ids if mid not in [movie_id]]
        return recommendations[:n_reccomend]

    # This method will recommend movies based on history stored in self.hist list
    # This is for those times when you know you have watched movies and based on all those movies, you want recommendation.
    # I am setting the function to recommend 5 movies
    def recommend_on_history(self,n_reccomend = 5):
        if self.ishist == False:
            return print('No history found')
        history = np.array([list(pivot_table.loc[mid]) for mid in self.hist])
        distance,neighbors = model.kneighbors([np.average(history,axis=0)],n_neighbors=n_reccomend + len(self.hist))
        movie_ids = [pivot_table.iloc[i].name for i in neighbors[0]]
        recommendations = [str(df_movies[df_movies['MovieId']==mid]['Title']).split('\n')[0].split('  ')[-1] for mid in movie_ids if mid not in self.hist]
        return recommendations[:n_reccomend]
```

```python
# Initializing the Recommender Object
recommender = Recommender()
```

```python
recommender.recommend_on_history()
```
